# Question 2:- In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
url=[]
url_links=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
for tag in url_links:
    if tag:
        text=tag.get_attribute('href')
        url.append(text)
    else:
        url.append("-") 
```

```python
brand_name=[]
name_product=[] 
price=[]
returne_exchange=[]
expected_delivery=[]
availability=[] 
product_URL=[] 
for i in url:
    driver.get(i)
    time.sleep(2)  
    #Extracting brand name
    try:
        brand_name_tags=driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-none"]')
        brand_name.append(brand_name_tags.text)
    except NoSuchElementException:
        brand_name.append('-')
      #Extracting product name   
    try:
        name_product_tags=driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        name_product.append(name_product_tags.text)
    except NoSuchElementException:
        name_product.append('-')
      #Extracting price   
    try:
        price_tags=driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        price.append(price_tags.text)
    except NoSuchElementException:
        price.append('-')  
      #Extracting return and exchange time
    try:
        returne_exchange_tags=driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]/span/div[2]/span')
        returne_exchange.append(returne_exchange_tags.text)
    except NoSuchElementException:
        returne_exchange.append('-')   
     #Extracting expected delivery  
    try:
        expected_delivery_tags=driver.find_element(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span[1]')
        expected_delivery.append(expected_delivery_tags.text)
    except NoSuchElementException:
        expected_delivery.append('-')
      #Extracting availability   
    try:
        availability_tags=driver.find_element(By.XPATH,'//*[@id="availability"]')
        availability.append(availability_tags.text)
    except NoSuchElementException:
        availability.append('-')
```

```python
len(brand_name),len(name_product),len(price),len(returne_exchange),len(expected_delivery),len(expected_delivery),len(availability)
```

```python
for _ in range(20):
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(2)  # Add a delay to allow comments to load

# Initialize lists to store comments, upvotes, and timestamps
comments = []
upvotes = []
timestamps = []

# Scrape comments, upvotes, and timestamps
while len(comments) < 500:
    comments_elements = driver.find_elements(By.XPATH, '//yt-formatted-string[@id="content-text"]')
    upvotes_elements = driver.find_elements(By.XPATH, '//span[@id="vote-count-middle"]')
    timestamps_elements = driver.find_elements(By.XPATH, '//a[@class="yt-simple-endpoint style-scope yt-formatted-string"]')

    # Extract text from elements
    for comment_element, upvote_element, timestamp_element in zip(comments_elements, upvotes_elements, timestamps_elements):
        comments.append(comment_element.text)
        upvotes.append(upvote_element.text)
        timestamps.append(timestamp_element.text)

    # Scroll down to load more comments if needed
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(2)  # Add a delay to allow comments to load

# Ensure we only have 500 comments, upvotes, and timestamps
comments = comments[:500]
upvotes = upvotes[:500]
timestamps = timestamps[:500]

# Print the number of comments scraped
print(f"Number of comments scraped: {len(comments)}")
    
```

```python
name=[]
ratings=[]
total_reviews=[]
overall_reviews=[]
private_price=[] 
dorms_price=[]
facilities=[]
property_description=[]
city_center=[]

#Extracting city center distance from hostel 
city_center_tags=driver.find_elements(By.XPATH,'//div[@class="property-distance"]')
for tags in city_center_tags:
    text=tags.text.replace('\n',' ')
    city_center.append(text)

for i in url:
    driver.get(i)
    time.sleep(2)
    
    #Extracting Title of hotel name
    try:
        name_tags=driver.find_element(By.XPATH,'//div[@class="title-2"]')
        name.append(name_tags.text)
    except NoSuchElementException:
        name.append('-')
        
    #Extracting Ratings of hotel name    
    try:
        ratings_tags=driver.find_element(By.XPATH,'//div[@class="score orange big"]')
        ratings.append(ratings_tags.text)
    except NoSuchElementException:
        ratings.append('-')  
        
    #Extracting Overall Reviews of hotel name
    try:
        overall_reviews_tags=driver.find_element(By.XPATH,'//div[@class="keyword"]')
        overall_reviews.append(overall_reviews_tags.text)
    except NoSuchElementException:
        overall_reviews.append('-')
     #Extracting Total Reviews of hotel name    
    try:
        total_reviews_tags=driver.find_element(By.XPATH,'//div[@class="reviews"]')
        total_reviews.append(total_reviews_tags.text)
    except NoSuchElementException:
        total_reviews.append('-')
        
     #Extracting private rooms price of hotel name
    try:
        private_price_tags=driver.find_element(By.XPATH,'//div[@class="rate-price-container"]')
        text1=private_price_tags.text.replace('%\n','% Discounted Price ').replace('\nRs',' Standard Price Rs.').replace('\n',' ')
        private_price.append(text1)
    except NoSuchElementException:
        private_price.append('-')    
     #Extracting dorms rooms price of hotel name
    try:
        dorms_price_tags=driver.find_element(By.XPATH,'//div[@class="rate-price-container"]')
        text2=dorms_price_tags.text.replace('%\n','% Discounted Price ').replace('\nRs',' Standard Price Rs.').replace('\n',' ')
        dorms_price.append(text2)
    except NoSuchElementException:
        dorms_price.append('-')
        
    #Extracting facilities of hotel name    
    try:
        facilities_tags=driver.find_element(By.XPATH,'//ul[@class="facility-sections"]')
        text=facilities_tags.text.replace('\n',' ')
        facilities.append(text)
    except NoSuchElementException:
        facilities.append('-') 
     #Extracting properties Discription of hotel name   
    try:
        property_description_tags=driver.find_element(By.XPATH,'//div[@class="content collapse-content"]')
        text=property_description_tags.text.replace('\n',' ')
        property_description.append(text)
    except NoSuchElementException:
        property_description.append('-')     
```
