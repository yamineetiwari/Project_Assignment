Question 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
page=driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
name = []
artist = []
upload_date = []
views = []

try:
    name_tags = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
    for tag in name_tags:
        text = tag.text
        text2 = re.sub(r'\[.*?\]', ' ', text)
        name.append(text2)
except NoSuchElementException:
    name.append('-')

# Extracting artist details from Wikipedia website
try:
    artist_tags = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
    for tag in artist_tags:
        artist.append(tag.text)
except NoSuchElementException:
    artist.append('-')

# Extracting upload date from Wikipedia website
try:
    upload_date_tags = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
    for tag in upload_date_tags:
        upload_date.append(tag.text)
except NoSuchElementException:
    upload_date.append('-')

# Extracting views from Wikipedia website
try:
    views_tags = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
    for tag in views_tags:
        views.append(tag.text)
except NoSuchElementException:
    views.append('-')
```

```python
len(name),len(artist),len(upload_date),len(views)
```

```python
df=pd.DataFrame({"NAME":name,"ARTIST":artist,"UPLOAD DATE":upload_date,"VIEWS":views})
df
```

```python
driver.close()
```

Question 4. Scrape the details of trending repositories on Github.com. Url = https://github.com/ You have to find the following details:

A) Repository title

B) Repository description

C) Contributors count

D) Language used

```python
driver.get("https://github.com/")
```

```python
#Entering Explore page
source_page=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
source_page.click()
```

```python
#Entering Trending page
trending_page=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
trending_page.click()
```

```python
#Extracting URL
url=[]
repository = driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
for i in repository:
    url.append(i.get_attribute("href"))
```

```python
lang=[]
des=[]
count=[]
title=[]
#Extracting Title of trending repositories
repos_title=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]')
for tags in repos_title:
    if tags:
        title.append(tags.text)
    else:
        title.append('-')
for i in url:
    driver.get(i)
    time.sleep(2)
    #Extracting description of trending repositories
    try:
        des_tags=driver.find_element(By.XPATH,'//p[@class="f4 my-3"]')
        des.append(des_tags.text)
    except NoSuchElementException:
        des.append('_')  
    #Extracting contribution count of trending repositories
    try:
        count_tags=driver.find_element(By.XPATH,'//span[@class="Counter ml-1"]')
        count.append(count_tags.text)
    except NoSuchElementException:
        count.append('_')  
   
    #Extracting language of trending repositories    
    try:
        lang_tags=driver.find_element(By.XPATH,'//ul[@class="list-style-none"]')
        text1=lang_tags.text.replace('%\n','%,').replace('\n','-')
        lang.append(text1)
    except NoSuchElementException:
        lang.append('_')
        
```

```python
len(title),len(des),len(count),len(lang)
```

```python
df=pd.DataFrame({"Title":title,"Repository description":des,"Contributors count":count,"Language Used":lang})
df
```

Question 8. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year
Note: - from the home page you have to go to the Show All Dataset page through code.

```python
driver.get("https://archive.ics.uci.edu/")
```

```python
#Entering into dataset and clicking dataset button
dataset_page=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
dataset_page.click()
```

```python
url=[]
url_link=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for link in url_link:
    text=link.get_attribute('href')
    url.append(text)
```

```python
dataset_name=[]
data_type=[]
task=[]
attri_type=[]
no_inst=[]
no_attri=[]
years=[]

for i in url:
    driver.get(i)
    time.sleep(2)
    #Extracting Dataset Name from UCI machine learning repositories 
    try:
        dataset_name_tags=driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        dataset_name.append(dataset_name_tags.text)
    except NoSuchElementException:
        dataset_name.append('-')

    #Extracting Data type from UCI machine learning repositories
    try:
        data_type_tags=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[2]/p')
        data_type.append(data_type_tags.text)
    except NoSuchElementException:
        data_type.append('-')
    #Extracting Tast from UCI machine Learning Repositories
    try:
        task_tags=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[3]/p')
        task.append(task_tags.text)
    except NoSuchElementException:
        task.append('-')
    #Extracting Attribute type from UCI machine Learning Repositories
    try:
        attri_type_tags=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[4]/p')
        attri_type.append(attri_type_tags.text)
    except NoSuchElementException:
        attri_type.append('-') 
    #Extracting no of instances from UCI machine Learning Repositories
    try:
        no_inst_tags=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[5]/p')
        no_inst.append(no_inst_tags.text)
    except NoSuchElementException:
        no_inst.append('-')
   #Extracting no of instances from UCI machine Learning Repositories
    try:
        no_attri_tags=driver.find_element(By.XPATH,'//div[@class="grid grid-cols-8 gap-4 md:grid-cols-12"]/div[6]/p')
        no_attri.append(no_attri_tags.text)
    except NoSuchElementException:
        no_attri.append('-')  
    #Extracting Year from UCI machine Learning Repositories
    try:
        year_tags = driver.find_element(By.XPATH, '//div[@class="relative flex flex w-full items-center gap-4 bg-primary p-2"]/div[2]/h2')
        text = year_tags.text
        match = re.search(r'\b\d{4}\b', text)  # Search for four consecutive digits representing the year
        if match:
            extracted_year = match.group()  # Extract the matched year
            years.append(extracted_year)  # Append the extracted year to the list
    except NoSuchElementException:
        years.append('-')  # Append a placeholder if the year cannot be extracted
```

```python
len(url),len(dataset_name),len(data_type),len(task),len(attri_type),len(no_inst),len(no_attri),len(years)
```

```python
df=pd.DataFrame({"Dataset Name":dataset_name,"Data Type":data_type,"Task":task,"Attribute Type":attri_type,"No of Instances":no_inst,"No of Attribute":no_attri,"Year":years})
df
```
