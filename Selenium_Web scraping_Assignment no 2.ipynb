scrap the mentioned data and make the DataFrame

```python
driver.get('https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1')
```

```python
#scraping the name of prime minister
name=[]
name_tags=driver.find_elements(By.XPATH,'//td[2][@valign="top"]/div')
for tags in name_tags:
    name.append(tags.text)

#scraping the born and dead of prime minister
born_dead=[]
born_dead_tags=driver.find_elements(By.XPATH,'//td[3][@valign="top"]/div')
for tags in born_dead_tags:
        born_dead.append(tags.text)
 
#scraping the born and dead of prime minister
term_of_office=[]
term_of_office_tags=driver.find_elements(By.XPATH,'//td[4][@valign="top"]')
for tags in term_of_office_tags:
    text=tags.text.strip().replace('\n',', ')
    term_of_office.append(text)
    
#scraping the born and dead of prime minister
term_of_office=[]
term_of_office_tags=driver.find_elements(By.XPATH,'//td[4][@valign="top"]')
for tags in term_of_office_tags:
    text=tags.text.strip().replace('\n',', ')
    term_of_office.append(text)
    
#Scraping the remarks of the prime minister    
remarks=[]
remarks_tags=driver.find_elements(By.XPATH,'//td[5][@valign="top"]/div')
for tags in remarks_tags:
        remarks.append(tags.text)

df=pd.DataFrame({"Name":name,"Term of Office":term_of_office,"born_dead":born_dead,"Remarks":remarks})
df        
```
